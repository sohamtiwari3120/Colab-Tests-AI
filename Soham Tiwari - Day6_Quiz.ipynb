# **Training**

```python
import numpy as np
```

```python
S=[0.6,0.5,0.3,0.2]
vLd=[10,9,7,4]
S=np.array(S)
vLd=np.array(vLd)
```

```python
def create_A(x,D):
  A=[]
  for i in x:
    temp=[]
    for d in range(0,D+1):
      temp+=[i**d]
    A+=[temp]
  return np.array(A)
```

```python
X_train=S
Yd=vLd
D=2
```

```python
def initialize_w(D):
  return np.random.random(D+1)
```

```python
def train_model(x,yd,D,eta):
  w=initialize_w(D)
  print("x =",x," ",x.shape)
  print(" ")
  print("yd =",yd," ",yd.shape)
  print(" ")
  print("w0 =",w," ",w.shape)
  A=create_A(x,D)
  print(" ")
  print("A =",A," ",A.shape)
  print(" ")
  E=np.zeros((x.shape[0],1))
  ctr=0
  for i in range(0,100):
    ctr+=1
    for j in range(0,x.shape[0]):
      a=A[j].reshape(D+1,1)
      w=w.reshape(D+1,1)
      y=a.T.dot(w)
      y=y[0][0]
      e=yd[j]-y
      e=np.array(e)
      E[j]=e**2
      e=e.reshape(1,1)
      w = w + eta*(a.dot(e))
  print("Counter =",ctr)
  return w,E
```

```python
w,E=train_model(S,vLd,D,0.06)
print("Final w =",w," ",w.shape)
print(" ")
print("Final error after 100 epochs E =",E," ",E.shape)
```
