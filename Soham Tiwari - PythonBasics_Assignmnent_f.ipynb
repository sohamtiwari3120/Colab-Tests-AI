Find whether a list a palindrome -> a sequence that reads the same forwards and backwards <br>
is_palindrome([1, 2, 1]) -> True <br>
is_palindrome([1, 2, 3, 2, 1] -> True <br>
is_palindrome([1, 2, 3, 1]) -> False

```python
def is_palindrome(l):
    """
    Input:
      l: Python list
    
    Output:
      boolean value indicating whether l is a palindrome or not
    
    """
    # YOUR CODE HERE
    rev=[]
    for i in l:
      rev=[i]+rev
    return l==rev
```

```python
"""Testing code for is_palindrome"""

assert(is_palindrome([1, 2, 1])) == True
assert(is_palindrome([1, 2, 3, 2, 1])) == True
assert(is_palindrome([1, 2, 3, 4])) == False

print("Sample Tests passed", '\U0001F44D')
```

Replace consecutive duplicate elements of list with single element. <br>
Eg., l = [a, a, a, b, b, c, a, a, d, d, d, x, x] <br>
compress(l) -> [a, b, c, a, d, x]

```python
def compress(l):
    """
    Input:
      l: Python list
    Output:
      Returns a list with consecutive duplicate elements replaced by a single element
    """
    # YOUR CODE HERE
    comp=[]
    flag=0
    j=0
    for i in range(0,len(l)):
      if(flag==0):
        comp+=[l[i]]
        flag=1
        continue
      if(comp[j]!=l[i]):
        comp+=[l[i]]
        j+=1
    return comp
      
    
```

```python
"""Testing code for compress"""

assert(compress([1, 2, 2])) == [1, 2]
assert(compress([1, 2, 2, 2, 1, 1, 3, 'x', 'x', 'x'])) == [1, 2, 1, 3, 'x']


print("Sample Tests passed", '\U0001F44D')
```

Pack consecutive duplicates of list elements into sublists. <br>
Eg., l = [1, 1, 1, 2, 2, 3, 3, 4]
pack(l) -> [[1,1,1], [2,2], [3,3], [4]]

```python
def pack(l):
    """
    Input:
      l: Python list
      
    Output:
      Returns a list with consecutive duplicate elements packed into sublists
    """
    # YOUR CODE HERE
    flag=0
    j=0
    p=[]
    temp=[]
    for i in range(0,len(l)):
      if(flag==0):
        temp+=[l[i]]
        flag=1
        continue
      if(l[i]==temp[j]):
        temp+=[l[i]]
        j+=1
      else:
        p+=[temp]
        temp=[l[i]]
        j=0
#         i=i-1
#         flag=0
    p+=[temp]
    return p
```

   split_list([],5)-->[]

```python
def split_list(l,k):
    """
    Input:
      l: python list
    k:
      Index at which to split l
    Output:
      single list if k > len(l)
      tuple of lists otherwise
    Returns splits list in the form of tuple. If k > len (l), return list as it is.
    """  
    # YOUR CODE HERE
    if(k>len(l)):
      return l
    else:
      return (l[0:k],l[k:])
```

  Flatten a nested list structure.
  Transform a list, possibly holding lists as elements into a `flat' list by replacing each          
  list with its elements (recursion)
   Example:
    def my_flatten([a, [b, [c, d], e]])-->[a, b, c, d, e]


```python
def my_flatten(l):
  """
  Input: python list
  Output:
    python list with nested lists flattened
  """
  # YOUR CODE HERE
  x=len(l)
  for i in range(0,x):
    if(type(l[i])==list):
      l=l[:i]+my_flatten(l[i])+l[i+1:]
      x=len(l)
  return l
```

```python
"""Testing code for flatten list"""
assert(my_flatten([1, 2, 3, 4,[]])) ==[1,2,3,4]
assert(my_flatten([1, 2,[3,4]])) == [1,2,3,4]


print("Sample Tests passed", '\U0001F44D')
```

 Example:<br>

```python
def is_prime(n):
  flag=0
  for i in range(2,n):
    if(n%i==0):
      flag=1
      break
      
  if flag==0:
    return True
  else:
    return False
```

 Determine the prime factors of a given positive integer<br>
 Construct a flat list containing the prime factors in ascending order.
 Example:<br>
  prime_factors(315)--> [3,3,5,7]

```python
import math
def prime_factors(n):
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
```

```python
print(prime_factors(12))
```

```python
"""Testing for prime factor of a number """
assert(prime_factors(12)) == [2,2,3]

print("Sample Tests passed", '\U0001F44D')
```

Write a Python program to combine two dictionary adding values for common keys.<br>
d1 = {'a': 100, 'b': 200, 'c':300}<br>
d2 = {'a': 300, 'b': 200, 'd':400}<br>Example-<br>
combine_duplicates(d1,d2)-->>({'a': 400, 'b': 400, 'd': 400, 'c': 300})

```python
def combine_duplicates(d1,d2):

  for i in d2.keys():
    if i in d1.keys():
      d1[i]+=d2[i]
    else:
      d1[i]=d2[i]
  return d1
    
```

```python
"""Test for combine_duplicates"""
assert combine_duplicates({'a':10, 'b':20}, {'a':50, 'c':20}) == {'a':60, 'b':20, 'c':20}
print("Sample Tests passed", '\U0001F44D')
```

Write a Python program to find the highest n values in a dictionary.<br>
Example-<br>
max_3({'a':500, 'b':5874, 'c': 560,'d':400, 'e':5000, 'f': 20}-->[5874,5000,500]

```python
def max_n(d,n):
  """
  Input:
    d: python dictionary
    n: int indicating number of highest values to return from d
  Output:
    list of length n containing highest n values of d. 
    (return all values if number of values in d is less than n)
  """
  # YOUR CODE HERE
  li=[]
  for i in d.keys():
    li+=[d[i]]
  return sorted(li,reverse=True)[:n]
```

```python
print( max_n({'a':500, 'b':5874, 'c': 560,'d':400, 'e':5000, 'f': 20},2))
```

```python
"""Test for max_n """
assert max_n({'a':500, 'b':5874, 'c': 560,'d':400, 'e':5000, 'f': 20},2) == [5874, 5000]
print("Sample Tests passed", '\U0001F44D')
```
